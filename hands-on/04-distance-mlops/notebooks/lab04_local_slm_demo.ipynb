# Lab 4 — Local SLM demo (optional)

Requires [Ollama](https://ollama.com) with a small model, e.g. `ollama pull llama3.2:1b`


```python
import json
try:
    import httpx
    prompt = "Summarize in one sentence: BGP peer down on router, customer escalated."
    r = httpx.post(
        "http://127.0.0.1:11434/api/generate",
        json={"model": "llama3.2:1b", "prompt": prompt, "stream": False},
        timeout=120.0,
    )
    r.raise_for_status()
    print(r.json().get("response", r.text))
except Exception as exc:
    print("Local SLM not available — skip or start Ollama:", exc)
```

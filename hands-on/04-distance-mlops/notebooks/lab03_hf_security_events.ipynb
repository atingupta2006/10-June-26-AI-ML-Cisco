# Lab 3 — HF: classify security event text


```python
from pathlib import Path

import pandas as pd

GH_ROOT = Path.cwd().resolve()
if GH_ROOT.name == "notebooks":
    GH_ROOT = GH_ROOT.parents[2]
else:
    for p in [GH_ROOT, *GH_ROOT.parents]:
        if (p / "data" / "support-ops").is_dir():
            GH_ROOT = p
            break
SUPPORT_OPS = GH_ROOT / "data" / "support-ops"

try:
    from transformers import pipeline
    _HF = True
except ImportError:
    _HF = False

sec = pd.read_csv(SUPPORT_OPS / "security_logs.csv")
label = sec["event_type"].iloc[0]
text = f"Security event type: {label}"
if _HF:
    clf = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
    result = clf(text, candidate_labels=["benign", "investigate", "critical"])
    print(result["labels"][0], result["scores"][0])
else:
    print("LABEL (keyword):", "investigate" if "deny" in label.lower() else "benign")
```

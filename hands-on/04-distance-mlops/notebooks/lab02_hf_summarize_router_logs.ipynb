# Lab 2 — HF: summarize router messages


```python
from pathlib import Path

import pandas as pd

GH_ROOT = Path.cwd().resolve()
if GH_ROOT.name == "notebooks":
    GH_ROOT = GH_ROOT.parents[2]
else:
    for p in [GH_ROOT, *GH_ROOT.parents]:
        if (p / "data" / "support-ops").is_dir():
            GH_ROOT = p
            break
SUPPORT_OPS = GH_ROOT / "data" / "support-ops"

try:
    from transformers import pipeline
    _HF = True
except ImportError:
    _HF = False

router = pd.read_csv(SUPPORT_OPS / "router_logs.csv")
text = router["message"].iloc[0]
if _HF:
    summarizer = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6")
    print(summarizer(text, max_length=35, min_length=8, do_sample=False)[0]["summary_text"])
else:
    print("SUMMARY (manual):", text[:100])
```

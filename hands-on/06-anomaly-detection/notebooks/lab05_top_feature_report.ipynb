# Lab 5 — Top feature report for leadership


```python
from pathlib import Path

import pandas as pd

GH_ROOT = Path.cwd().resolve()
if GH_ROOT.name == "notebooks":
    GH_ROOT = GH_ROOT.parents[2]
else:
    for p in [GH_ROOT, *GH_ROOT.parents]:
        if (p / "data" / "support-ops").is_dir():
            GH_ROOT = p
            break
SUPPORT_OPS = GH_ROOT / "data" / "support-ops"

import json

tickets = pd.read_csv(SUPPORT_OPS / "support_tickets.csv")
top = tickets["feature"].value_counts().head(5)
report = {
    "top_features": top.to_dict(),
    "total_tickets": int(len(tickets)),
    "narrative": f"Lead with '{top.index[0]}' — {int(top.iloc[0])} tickets in sample.",
}
out = GH_ROOT / "hands-on" / "06-anomaly-detection" / "output" / "top_features_report.json"
out.parent.mkdir(parents=True, exist_ok=True)
out.write_text(json.dumps(report, indent=2), encoding="utf-8")
print(json.dumps(report, indent=2))
```

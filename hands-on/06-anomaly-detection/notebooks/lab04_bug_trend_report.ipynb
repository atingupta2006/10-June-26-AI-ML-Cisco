# Lab 4 — Bug trend report


```python
from pathlib import Path

import pandas as pd

GH_ROOT = Path.cwd().resolve()
if GH_ROOT.name == "notebooks":
    GH_ROOT = GH_ROOT.parents[2]
else:
    for p in [GH_ROOT, *GH_ROOT.parents]:
        if (p / "data" / "support-ops").is_dir():
            GH_ROOT = p
            break
SUPPORT_OPS = GH_ROOT / "data" / "support-ops"

import matplotlib.pyplot as plt

tickets = pd.read_csv(SUPPORT_OPS / "support_tickets.csv", parse_dates=["created_date"])
weekly = tickets.set_index("created_date").resample("W").size()
wow = weekly.pct_change().iloc[-1] * 100
print(f"Latest week tickets: {int(weekly.iloc[-1])}")
print(f"Week-over-week change: {wow:.1f}%")
weekly.plot(title="Bug ticket trend")
plt.tight_layout()
out = GH_ROOT / "hands-on" / "06-anomaly-detection" / "output" / "bug_trend_report.png"
out.parent.mkdir(parents=True, exist_ok=True)
plt.savefig(out)
plt.show()
```

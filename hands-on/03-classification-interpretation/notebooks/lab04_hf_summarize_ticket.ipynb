# Lab 4 — Hugging Face: summarize ticket text

Uses a small summarization pipeline (first run downloads weights).


```python
from pathlib import Path

import pandas as pd

GH_ROOT = Path.cwd().resolve()
if GH_ROOT.name == "notebooks":
    GH_ROOT = GH_ROOT.parents[2]
else:
    for p in [GH_ROOT, *GH_ROOT.parents]:
        if (p / "data" / "support-ops").is_dir():
            GH_ROOT = p
            break
SUPPORT_OPS = GH_ROOT / "data" / "support-ops"

try:
    from transformers import pipeline
    _HF = True
except ImportError:
    _HF = False
    print("SKIP: pip install transformers torch (class VM; first run downloads models)")

tickets = pd.read_csv(SUPPORT_OPS / "support_tickets.csv")
text = tickets["summary"].iloc[0]
print("INPUT:", text)
if _HF:
    summarizer = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6")
    out = summarizer(text, max_length=40, min_length=8, do_sample=False)
    print("SUMMARY:", out[0]["summary_text"])
else:
    print("SUMMARY (manual):", text[:120] + ("..." if len(text) > 120 else ""))
```

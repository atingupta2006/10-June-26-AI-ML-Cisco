> **Checkpoint:** all **six** phases in order · activity + deliverable per phase

**Format:** Worksheet / group discussion (no verification script)

---


## Why this matters

CRISP-DM is the **shared language** between business sponsors and data teams. When a stakeholder asks for "AI," mapping their ask to a phase prevents skipping data quality work or deploying models without evaluation.



## CRISP-DM lifecycle

```text
Business Understanding → Data Understanding → Data Preparation
        → Modeling → Evaluation → Deployment
```

Not every project needs **Modeling** (e.g. Excel pivot answers many business questions).

---

## 1. Six phases (reference)

````python
CRISP_DM_PHASES = [
    "Business Understanding",
    "Data Understanding",
    "Data Preparation",
    "Modeling",
    "Evaluation",
    "Deployment",
]

for i, phase in enumerate(CRISP_DM_PHASES, start=1):
    print(f"{i}. {phase}")
````

---

## 2. Worked example — team sales (Day 1 capstone preview)

````python
import pandas as pd
from IPython.display import display

business_question = "Which region had the highest Q2 sales?"

team_sales_example = pd.DataFrame({
    "phase": CRISP_DM_PHASES,
    "activity": [
        "Define regional performance KPI",
        "Open team_sales.csv; profile 100 teams, 4 regions",
        "Ensure numeric columns; handle blanks",
        "Optional forecast (not required Day 1)",
        "Compare regional Q2 totals; count growth teams",
        "Share chart with manager / stakeholders",
    ],
    "deliverable": [
        "Problem statement",
        "EDA summary",
        "Analysis-ready table",
        "Model artifact (if used)",
        "Regional ranking answer",
        "Presentation or email report",
    ],
    "needs_ml": ["No", "No", "No", "Optional", "No", "No"],
})

print(f"Business question: {business_question}")
display(team_sales_example)
````

Excel pivot / `SUMIF` in Lab 6 fits **Data Preparation → Evaluation** — DS without ML.

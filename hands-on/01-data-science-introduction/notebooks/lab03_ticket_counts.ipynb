# Lab 3 — Ticket counts by product

**Quick check:** 800 tickets · 6 products


```python
from pathlib import Path

import pandas as pd

GH_ROOT = Path.cwd().resolve()
if GH_ROOT.name == "notebooks":
    GH_ROOT = GH_ROOT.parents[2]
else:
    for p in [GH_ROOT, *GH_ROOT.parents]:
        if (p / "data" / "support-ops").is_dir():
            GH_ROOT = p
            break
SUPPORT_OPS = GH_ROOT / "data" / "support-ops"

tickets = pd.read_csv(SUPPORT_OPS / "support_tickets.csv")
print("shape:", tickets.shape)
print(tickets["product"].value_counts())
print("mean severity high+critical:", tickets["severity"].isin(["high", "critical"]).mean())
```

(e.g. calling a SQL dashboard "machine learning").

```
        +-------------------------------------+
        |  Artificial Intelligence (AI)       |
        |   +-----------------------------+   |
        |   |  Machine Learning (ML)      |   |
        |   |    +-----------------+    |   |
        |   |    | Data Science    |    |   |
        |   |    | (stats + code)  |    |   |
        |   |    +-----------------+    |   |
        |   +-----------------------------+   |
        +-------------------------------------+
```

**Instructor hint:** Netflix recommendations sit in **ML**; a quarterly sales pivot in Excel is **DS** without ML.

## The nesting mental model

```text
┌─────────────────────────────────────────────┐
│  ARTIFICIAL INTELLIGENCE (AI)               │
│  ┌───────────────────────────────────────┐  │
│  │  MACHINE LEARNING (ML)                │  │
│  │     ┌─────────────────────────────┐   │  │
│  │     │  DATA SCIENCE (DS)          │   │  │
│  │     │  (overlaps ML, extends out) │   │  │
│  │     └─────────────────────────────┘   │  │
│  └───────────────────────────────────────┘  │
└─────────────────────────────────────────────┘
```

| Term | One-line definition |
|------|---------------------|
| **AI** | Systems that perform tasks requiring human-like intelligence |
| **ML** | Algorithms that learn patterns from data without explicit rules |
| **Data Science** | End-to-end process of extracting insight from data |

## 4. Optional — visual Venn diagram

````python
%matplotlib inline

import matplotlib.pyplot as plt
from matplotlib.patches import Circle

fig, ax = plt.subplots(figsize=(7, 5))
ai = Circle((0.5, 0.5), 0.42, fill=False, linewidth=2, color="steelblue")
ml = Circle((0.45, 0.5), 0.28, fill=False, linewidth=2, color="seagreen")
ds = Circle((0.55, 0.48), 0.22, fill=False, linewidth=2, color="coral")
for patch in (ai, ml, ds):
    ax.add_patch(patch)
ax.text(0.5, 0.92, "AI", ha="center", fontsize=12, color="steelblue")
ax.text(0.38, 0.62, "ML", ha="center", fontsize=11, color="seagreen")
ax.text(0.62, 0.55, "DS", ha="center", fontsize=11, color="coral")
ax.text(0.42, 0.45, "Netflix\nrecs", ha="center", fontsize=9)
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect("equal")
ax.axis("off")
ax.set_title("AI ⊃ ML; Data Science overlaps ML")
plt.tight_layout()
plt.show()
````

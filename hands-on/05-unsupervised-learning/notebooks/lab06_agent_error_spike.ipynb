# Lab 6 — Agentic case B: router error spike


```python
from pathlib import Path

import pandas as pd

GH_ROOT = Path.cwd().resolve()
if GH_ROOT.name == "notebooks":
    GH_ROOT = GH_ROOT.parents[2]
else:
    for p in [GH_ROOT, *GH_ROOT.parents]:
        if (p / "data" / "support-ops").is_dir():
            GH_ROOT = p
            break
SUPPORT_OPS = GH_ROOT / "data" / "support-ops"

def agent_spike() -> str:
    r = pd.read_csv(SUPPORT_OPS / "router_logs.csv")
    top = r["error_code"].value_counts().head(1)
    code, count = top.index[0], int(top.iloc[0])
    return (
        f"Spike check: {code} appears {count} times in sample. "
        f"Recommend: open bridge call if {code} in ('E-BGP','E-500'). "
        f"Attach top 5 messages to ticket."
    )

print(agent_spike())
```

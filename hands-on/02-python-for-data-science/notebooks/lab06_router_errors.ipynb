# Lab 6 — Router error code trend


```python
from pathlib import Path

import pandas as pd

GH_ROOT = Path.cwd().resolve()
if GH_ROOT.name == "notebooks":
    GH_ROOT = GH_ROOT.parents[2]
else:
    for p in [GH_ROOT, *GH_ROOT.parents]:
        if (p / "data" / "support-ops").is_dir():
            GH_ROOT = p
            break
SUPPORT_OPS = GH_ROOT / "data" / "support-ops"

import matplotlib.pyplot as plt

router = pd.read_csv(SUPPORT_OPS / "router_logs.csv", parse_dates=["timestamp"])
top = router["error_code"].value_counts().head(8)
top.plot(kind="bar", figsize=(7, 3), title="Top router error codes")
plt.tight_layout()
plt.show()
```

# Lab 5 — Bug ticket trend (weekly)

**Trend analysis** client use case.


```python
from pathlib import Path

import pandas as pd

GH_ROOT = Path.cwd().resolve()
if GH_ROOT.name == "notebooks":
    GH_ROOT = GH_ROOT.parents[2]
else:
    for p in [GH_ROOT, *GH_ROOT.parents]:
        if (p / "data" / "support-ops").is_dir():
            GH_ROOT = p
            break
SUPPORT_OPS = GH_ROOT / "data" / "support-ops"

import matplotlib.pyplot as plt

tickets = pd.read_csv(SUPPORT_OPS / "support_tickets.csv", parse_dates=["created_date"])
weekly = tickets.set_index("created_date").resample("W").size()
print(weekly.tail())
weekly.plot(figsize=(8, 3), title="Tickets per week")
plt.ylabel("ticket count")
plt.tight_layout()
out = GH_ROOT / "hands-on" / "02-python-for-data-science" / "output" / "bug_trend.png"
plt.savefig(out)
plt.show()
print("saved", out.name)
```
